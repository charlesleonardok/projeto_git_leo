```python
#Importando bibliotecas e criando a função para plotar e calcular
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

def plot_and_calculate(z_min, z_max=None, direction="less"):
    fig, ax = plt.subplots()
    ax.plot(x, y)

    if z_max is None:
        if direction == ">":
            prob = 1 - norm.cdf(z_min)
            x_fill = np.linspace(z_min, 4, 1000)
            title = f'P(Z > {z_min}) = {prob:.4f}'
        elif direction == "<":
            prob = norm.cdf(z_min)
            x_fill = np.linspace(-4, z_min, 1000)
            title = f'P(Z < {z_min}) = {prob:.4f}'
        else:
            raise ValueError("Invalid direction. Choose '>' or '<'.")
    else:
        prob = norm.cdf(z_max) - norm.cdf(z_min)
        x_fill = np.linspace(z_min, z_max, 1000)
        title = f'P({z_min} < Z < {z_max}) = {prob:.4f}'

    y_fill = norm.pdf(x_fill, 0, 1)
    ax.fill_between(x_fill, y_fill, alpha=0.5, label=title)
    ax.set_title(title)
    ax.legend()
    plt.show()
```

```python
plot_and_calculate(-1.25,direction=">")  # Para P(Z > -1.25)
```

```python
plot_and_calculate(-1.5, 2.53)  # Para P(-1.5 < Z < 2.53)
```
